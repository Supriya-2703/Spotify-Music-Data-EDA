#### Preprocessing:

```python
import pandas as pd
```

```python
df = pd.read_csv("dataset.csv")
df.head()
```

```python
df.info()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df.columns
```

```python
essential_cols = ["tempo", "popularity", "track_genre"]
df = df.dropna(subset=essential_cols)
```

```python
numeric_cols = [
    "tempo", "popularity", "duration_ms", "danceability", "energy",
    "loudness", "valence", "speechiness", "acousticness",
    "instrumentalness", "liveness", "time_signature", "year"
]
```

```python
for col in numeric_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")
```

```python
numeric_cols_existing = [col for col in numeric_cols if col in df.columns]

df[numeric_cols_existing].isnull().sum()
```

```python
df = df.dropna(subset=["tempo", "popularity"])
```

```python
df = df[df["duration_ms"] > 0]
```

```python
df = df[(df["tempo"] > 40) & (df["tempo"] < 250)]
```

```python
df = df[(df["popularity"] >= 0) & (df["popularity"] <= 100)]
```

```python
df["duration_min"] = df["duration_ms"] / 60000
```

```python
df["main_genre"] = df["track_genre"].str.split(" ").str[0]
```

```python
df.info()
```

```python
df.head()
```

```python
df.describe()
```

```python
df.to_csv("dataset_cleaned.csv", index=False)
```

#### 1.Distribution of TEMPO

```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
import numpy as np
```

```python
df = df.dropna(subset=["tempo", "popularity", "track_genre"])
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["tempo"], bins=40)
plt.xlabel("Tempo (BPM)")
plt.ylabel("Count")
plt.title("Distribution of Tempo")
plt.show()
```

#### 2.Distribution of POPULARITY

```python
plt.figure(figsize=(8,5))
plt.hist(df["popularity"], bins=30)
plt.xlabel("Popularity (0–100)")
plt.ylabel("Count")
plt.title("Distribution of Popularity")
plt.show()
```

#### 3.Genre-wise Popularity Comparison

```python
genre_medians = df.groupby("track_genre")["popularity"].median().sort_values(ascending=False)
top_genres = genre_medians.head(10).index

df_top = df[df["track_genre"].isin(top_genres)]

genre_groups = [df_top[df_top["track_genre"] == g]["popularity"] for g in top_genres]

plt.figure(figsize=(12,6))
plt.boxplot(genre_groups, labels=top_genres, showfliers=False)  # hide outliers for cleaner plot
plt.xticks(rotation=45)
plt.ylabel("Popularity")
plt.title("Popularity by Genre (Top 10 Genres, Sorted by Median Popularity)")
plt.tight_layout()
plt.show()
```

#### 4.Correlation Heatmap

```python
features = [
    "tempo", "danceability", "energy", "loudness",
    "valence", "duration_ms", "popularity"
]

features_available = [f for f in features if f in df.columns]

corr = df[features_available].corr()

plt.figure(figsize=(8,6))
plt.imshow(corr, cmap="viridis")
plt.colorbar()
plt.xticks(range(len(features_available)), features_available, rotation=45)
plt.yticks(range(len(features_available)), features_available)
plt.title("Correlation Heatmap")
plt.show()
```

#### 5.Tempo vs Popularity

```python
x = df["tempo"]
y = df["popularity"]

m, b = np.polyfit(x, y, 1)

plt.figure(figsize=(8,6))
plt.scatter(x, y, alpha=0.4)
plt.plot(x, m*x + b)
plt.xlabel("Tempo (BPM)")
plt.ylabel("Popularity")
plt.title("Tempo vs Popularity (with Regression Line)")
plt.show()
```

#### Pairwise Scatter Plots for Key Features

```python
important_features = ["danceability", "energy", "tempo", "loudness", "valence"]

for col in important_features:
    if col in df.columns:
        plt.figure(figsize=(6,4))
        plt.scatter(df[col], df["popularity"])
        plt.xlabel(col)
        plt.ylabel("Popularity")
        plt.title(f"{col} vs Popularity")
        plt.show()
```
